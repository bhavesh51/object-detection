```python
import pandas as pd

for i in csv_paths:
    data=pd.read_csv(i)
    for index, row in data.iterrows():
        # print(row[0])
        width = row[1]
        height = row[2]
        class_name = row[3]
        xmin = row[4]
        ymin = row[5]
        xmax = row[6]
        ymax = row[7]
        x_center = 0.5*(xmin+xmax)
        y_center = 0.5*(ymin+ymax)
        b_width = xmax - xmin
        b_height= ymax - ymin
        class_num = np.where(classes==class_name)[0][0]
        # print(class_num,x_center,y_center,b_width,b_height)
        file_name ='{0}.txt'.format(row[0]) 
        with open(file_name, 'w') as f:
            f.write(str(class_num))
            f.write(' ')
            f.write(str(x_center/width))
            f.write(' ')
            f.write(str(y_center/height))
            f.write(' ')
            f.write(str(b_width/width))
            f.write(' ')
            f.write(str(b_height/height))
            f.close()  
```

```python
from tqdm import tqdm
import shutil
def splitting_data(imgs,anots,dest_imgs,drst_anots):
    for idx, (i, j) in tqdm(enumerate(zip(imgs, anots)),total=len(imgs)):
        shutil.copy(i,dest_imgs)
        shutil.copy(j,drst_anots)

splitting_data(X_train,y_train,'/home/bhavesh/Desktop/Aircraft/aircraft/images/train', '/home/bhavesh/Desktop/Aircraft/aircraft/labels/train')
splitting_data(X_test, y_test, '/home/bhavesh/Desktop/Aircraft/aircraft/images/valid', '/home/bhavesh/Desktop/Aircraft/aircraft/labels/valid')
```
